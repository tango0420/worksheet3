```python
#df['Age']=df['Age'].median())

#3.2 Exploratory Data Analysis Practice Exercise - 1.
# Question 1 to 5-
df['Fare_per_Year'] = df['Fare'] / df['Age']
print(df[['Fare', 'Age', 'Fare_per_Year']].head())


high_fare_age = df[df['Fare_per_Year'] > 5]
print(high_fare_age.head())


high_fare_age_srt = high_fare_age.sort_values(by='Fare_per_Year', ascending=False)
print(high_fare_age_srt.head())


result = high_fare_age_srt[['Name', 'Fare_per_Year']]
print(result.head())

```

```python
#Which adult male passenger (age â‰¥ 18 and Sex is â€™maleâ€™) paid the highest fare relative to their class?
#Question number 1 to 5-

df['Fare_per_Class'] = df['Fare'] / df['Pclass']
print(df[['Fare', 'Pclass', 'Fare_per_Class']].head())


adult_males = df[(df['Sex'] == 'male') & (df['Age'] >= 18)]
print(adult_males.head())


adult_males_srt = adult_males.sort_values(by='Fare_per_Class', ascending=False)
print(adult_males_srt.head())


result = adult_males_srt[['Name', 'Age', 'Fare_per_Class']]
print(result.head())
```

```python
#3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:
#Question 1 to 5-

total_fare = df['Fare'].sum()
print(total_fare)


first_class_passengers = df[df['Pclass'] == 1]
total_fare_first_class = first_class_passengers['Fare'].sum()
print(total_fare_first_class)

second_class_passengers = df[df['Pclass'] == 2]
total_fare_second_class = second_class_passengers['Fare'].sum()
print(total_fare_second_class)

third_class_passengers = df[df['Pclass'] == 3]
total_fare_third_class = third_class_passengers['Fare'].sum()
print(total_fare_third_class)


```

```python
#What percent of the total number of passengers on the Titanic belonged to each age group (e.g., child, adult, senior)?
#Question 1 to 5-

df['Age_group'] = pd.cut(df['Age'], bins=[0, 18, 64, float('inf')], labels=['child', 'adult', 'senior'], right=False)
print(df[['Age', 'Age_group']].head())


total_passengers = df.shape[0]
print(total_passengers)


age_group_counts = df['Age_group'].value_counts()
print(age_group_counts)


age_group_proportions = age_group_counts / total_passengers
print(age_group_proportions)


age_group_proportions_percentage = age_group_proportions * 100
print(age_group_proportions_percentage)
```
